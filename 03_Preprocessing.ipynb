```python
import pandas as pd
import numpy as np
```

```python
# Load data
matches_df = pd.read_csv("./data/matches.csv", encoding='utf-8')
matches_df = matches_df.drop(['Unnamed: 0'], axis=1)
hitter_df = pd.read_csv("./data/hitter.csv", encoding='utf-8')
pitcher_df = pd.read_csv("./data/pitcher.csv", encoding='utf-8')
```

```python
matches_df.tail()
```

## 1) pitcher_column

- ERA: 평균자책점
- WHIP: 이닝당 출루허용률
- H: 피안타
- HR: 홈런
- SO: 삼진
- BB: 볼넷
- HBP: 사구

```python
pitcher_df['h/g'] = pitcher_df.H / pitcher_df.G
pitcher_df['hr/g'] = pitcher_df.HR / pitcher_df.G
pitcher_df['so/g'] = pitcher_df.SO / pitcher_df.G
pitcher_df['bb/g'] = pitcher_df.BB / pitcher_df.G
pitcher_df['hbp/g'] = pitcher_df.HBP / pitcher_df.G
```

```python
matches_df = matches_df.rename(columns={
    'pitcher_home': 'name'
})
```

```python
target_columns = ['name', 'ERA', 'WHIP', 'h/g', 'hr/g', 'so/g', 'bb/g', 'hbp/g']
```

```python
matches_df1 = matches_df.merge(pitcher_df.ix[:, target_columns], on='name', how='inner')
matches_df1.tail()
```

```python
matches_df1 = matches_df1.rename(columns={
    'pitcher_away': 'name1'
})
```

```python
pitcher_df1 = pitcher_df.rename(columns={
    'name': 'name1'
})
```

```python
target_columns = ['name1', 'ERA', 'WHIP', 'h/g', 'hr/g', 'so/g', 'bb/g', 'hbp/g']
```

```python
matches_df2 = matches_df1.merge(pitcher_df1.ix[:, target_columns], on='name1', how='inner')
matches_df2.tail()
```

```python
drop_column_list = ['home_team', 'name', 'hitter_home', 
                    'away_team', 'name1', 'hitter_away']
```

```python
matches_df3 = matches_df2.drop(drop_column_list, axis=1)
```

```python
matches_df3.to_csv('./data/model_test.csv', encoding='utf-8')
```
